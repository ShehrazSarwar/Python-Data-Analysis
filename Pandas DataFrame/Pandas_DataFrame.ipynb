### What is a DataFrame?
A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

```python
import numpy as np
import pandas as pd
```

#### Creating DataFrame

```python
# using lists
student_data = [
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,50,2]
]
pd.DataFrame(student_data, columns = ['iq','marks','package'])
```

```python
# using dict
student_dict = {
    'name': ['Umer','Hassan','Shehraz','Fahad','Ahsan','Wasi'],
    'iq': [100,80,120,80,0,0],
    'marks':[80,70,100,50,0,0],
    'package': [10,7,14,2,0,0]
}

students = pd.DataFrame(student_dict)
students.set_index('name',inplace = True) 
students
```

#### Selecting rows from a DataFrame
<b> iloc: </b> Position based indexing <br>
<b> loc:  </b> Label based indexing

```python
# single row
movies.iloc[5]
```

```python
# multiple rows
movies.iloc[5:16:3] # same python slicing
```

```python
# fancy indexing
movies.iloc[[0,4,5]]
```
