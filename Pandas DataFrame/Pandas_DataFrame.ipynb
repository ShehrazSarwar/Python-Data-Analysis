### What is a DataFrame?
A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

#### Creating DataFrame

```python
# using lists
student_data = [
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,50,2]
]
pd.DataFrame(student_data, columns = ['iq','marks','package'])
```

```python
# using dict
student_dict = {
    'name': ['Umer','Hassan','Shehraz','Fahad','Ahsan','Wasi'],
    'iq': [100,80,120,80,0,0],
    'marks':[80,70,100,50,0,0],
    'package': [10,7,14,2,0,0]
}

students = pd.DataFrame(student_dict)
students.set_index('name',inplace = True) 
students
```

#### Filtering a DataFrame

```python
ipl.head(2)
```

```python
# find all the final winners from the data set of ipl
mask = ipl['MatchNumber'] == 'Final'
new_df = ipl[ipl['MatchNumber'] == 'Final']
new_df[['Season','WinningTeam']]

# in one line
ipl[ipl['MatchNumber'] == 'Final'][['Season','WinningTeam']]
```

```python
# how many super over finishes have occured in ipl
ipl[ipl['SuperOver'] == 'Y']['SuperOver'].count()
# or
ipl[ipl['SuperOver'] == 'Y'].shape[0]
# or
(ipl['SuperOver'] == 'Y').sum()
```

```python
# how many matches has csk won in Kolkata
# direct using boolean and sum
((ipl['City'] == 'Kolkata') & (ipl['WinningTeam'] == 'Chennai Super Kings')).sum()
# 2nd method boolean indexing
ipl[(ipl['City'] == 'Kolkata') & (ipl['WinningTeam'] == 'Chennai Super Kings')].shape[0]
```

```python
# fetch chances of toss winner is a match winner also in percentage
toss_winner = (ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0] / ipl.shape[0]) * 100
print(round(toss_winner,2),'%')
```

```python
# movies with rating higher than 8 and votes > 10000
len(movies[(movies['imdb_rating'] > 8) & (movies['imdb_votes'] > 10000)]) # or .shape
```
