Applying builtin aggregation functions on groupby objects

```python
# default numeric_only = False, so it will concatenate strings in case of sum()
genres.sum(numeric_only = True)
```

```python
genres.min()
```

```python
# all are applicable like std(), mode(), var(), prod() etc
genres.mean(numeric_only = True)
```

```python
# find the top 3 genres by total earning
movies.groupby('Genre').sum(numeric_only=True)['Gross'].sort_values(ascending = False).head(3)
```

```python
movies.groupby('Genre')['Gross'].sum().sort_values(ascending = False).head(3) # efficient one
```

```python
# find the genre with highest avg IMDB rating
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending = False).head(1)
```

```python
# find director with most popularity
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending = False).head(1)
```

```python
# find the highest rated movie of each genre
highest_rated = movies.loc[movies.groupby('Genre')['IMDB_Rating'].idxmax()]
highest_rated[['Genre','Series_Title','IMDB_Rating']].sort_values(by = 'IMDB_Rating', ascending = False)
```

```python
# without groupby()
movies.sort_values(by = 'IMDB_Rating', ascending = False).drop_duplicates(subset = 'Genre')[['Genre','Series_Title','IMDB_Rating']]
```

```python
# find the no of movies done by each actor
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)
```

```python
# agg method
# passing dict
genres.agg(
    {
        'Runtime':'mean',
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Metascore':'max'
    }
)
```

```python
# passing list
# Only works if all columns have numeric data
# genres.agg(['min','max','mean'])
```

```python
# Using both
genres.agg(
    {
        'Runtime':['min','max','mean'],
        'IMDB_Rating':['min','max','mean'],
        'No_of_Votes':['min','max','mean'],
        'Gross':'sum',
        'Metascore':['min','max']
    }
)
```

```python
# looping on groups
for group,data in genres:
  print(type(group),type(data))
```

```python
# 1st Method:
df = pd.DataFrame()

for group, data in genres:
    top_movies = data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]
    df = pd.concat([df, top_movies])

df
```

```python
# 2nd Method (efficient): appending all dataframes to list and in the end concate the list
rows = []

for group, data in genres:
    rows.append(data[data['IMDB_Rating'] == data['IMDB_Rating'].max()])


df = pd.concat(rows, ignore_index = True) # ignore_index = False by default
df
```

```python
# split (apply) combine
# apply --> builtin function
```

```python
# find number of movies starting with A for each group
def foo(group):
  return group['Series_Title'].str.startswith('A').sum()


print(genres.apply(foo, include_groups = False))
```

```python
# find ranking of each movie in the group according to IMDB Score
def rank_movie(group):
  group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False, method = 'min')
  return group

new_df = genres.apply(rank_movie, include_groups = False).reset_index().set_index('level_1').rename_axis(index = None).sort_index()
col = new_df.pop('Genre')
new_df.insert(new_df.columns.get_loc('IMDB_Rating'), 'Genre', col)
```
